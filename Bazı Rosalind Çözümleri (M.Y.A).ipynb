```python
def rabbit_pairs(n, k):
    if n == 1 or n == 2:
        return 1
    else:
        return rabbit_pairs(n-1, k) + k * rabbit_pairs(n-2, k)

n = 30
k = 3
total_pairs = rabbit_pairs(n, k)
print(total_pairs)
```

```python
def highest_GC_content(file):
    sequences = {}
    current_id = None
    current_seq = ''
    
    with open(file, 'r') as f:
        for line in f:
            if line.startswith('>'):
                # yeni bir dizinin başlangıcı
                if current_id is not None:
                    # önceki diziyi saklayın
                    sequences[current_id] = current_seq
                current_id = line[1:].strip()
                current_seq = ''
            else:
                current_seq += line.strip()
        
        # son diziyi saklayın
        sequences[current_id] = current_seq
    
    gc_contents = {}
    
    for id, seq in sequences.items():
        gc_count = seq.count('G') + seq.count('C')
        gc_content = gc_count / len(seq) * 100
        gc_contents[id] = gc_content
    
    highest_id = max(gc_contents, key=gc_contents.get)
    highest_gc_content = gc_contents[highest_id]
    
    return highest_id, highest_gc_content

file = 'rosalind_gc.txt'
highest_id, highest_gc_content = highest_GC_content(file)
print(highest_id)
print(round(highest_gc_content, 6))
```

```python
def hamming_distance(file):
    with open(file, 'r') as f:
        s = f.readline().strip()
        t = f.readline().strip()
    
    distance = 0
    
    for i in range(len(s)):
        if s[i] != t[i]:
            distance += 1
    
    return distance

file = 'rosalind_hamm.txt'
hamming_distance(file)
```

```python
k, m, n = map(int, input().split())

total_pop = k + m + n

# Baskın alele sahip yavruların olasılığını hesaplayın
# P(Dominant alel) = 1 - P(Resesif alel)
# P(Resesif alel) = P(Homozigot resesif) + P(Heterozigot) * P(Resesif alel)
# P(Baskın alel) = 1 - P(Homozigot resesif) - P(Heterozigot) * P(Resesif alel)

prob_hom_rec = (n / total_pop) * ((n - 1) / (total_pop - 1))
prob_het = (m / total_pop) * ((m - 1) / (total_pop - 1))
prob_rec = (m / total_pop) * (n / (total_pop - 1)) + (n / total_pop) * (m / (total_pop - 1))
prob_dom = 1 - prob_hom_rec - prob_het * 0.25 - prob_rec * 0.5

print(prob_dom)
```

```python
# RNA codon table
codon_table = {
    "UUU": "F", "UUC": "F", "UUA": "L", "UUG": "L",
    "UCU": "S", "UCC": "S", "UCA": "S", "UCG": "S",
    "UAU": "Y", "UAC": "Y", "UAA": "Stop", "UAG": "Stop",
    "UGU": "C", "UGC": "C", "UGA": "Stop", "UGG": "W",
    "CUU": "L", "CUC": "L", "CUA": "L", "CUG": "L",
    "CCU": "P", "CCC": "P", "CCA": "P", "CCG": "P",
    "CAU": "H", "CAC": "H", "CAA": "Q", "CAG": "Q",
    "CGU": "R", "CGC": "R", "CGA": "R", "CGG": "R",
    "AUU": "I", "AUC": "I", "AUA": "I", "AUG": "M",
    "ACU": "T", "ACC": "T", "ACA": "T", "ACG": "T",
    "AAU": "N", "AAC": "N", "AAA": "K", "AAG": "K",
    "AGU": "S", "AGC": "S", "AGA": "R", "AGG": "R",
    "GUU": "V", "GUC": "V", "GUA": "V", "GUG": "V",
    "GCU": "A", "GCC": "A", "GCA": "A", "GCG": "A",
    "GAU": "D", "GAC": "D", "GAA": "E", "GAG": "E",
    "GGU": "G", "GGC": "G", "GGA": "G", "GGG": "G",
}

rna_string = input().strip()

protein_string = ""
for i in range(0, len(rna_string), 3):
    codon = rna_string[i:i+3]
    amino_acid = codon_table[codon]
    if amino_acid == "Stop":
        break
    protein_string += amino_acid

print(protein_string)
```
